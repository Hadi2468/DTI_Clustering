```python
import matplotlib.pyplot as plt
def show_slices(slices):
    fig, axes = plt.subplots(1, len(slices), figsize=(15,5))
    for i, slice in enumerate(slices):
#         axes[i].imshow(slice.T, cmap="gray", origin="upper")
        axes[i].imshow(slice.T, cmap="Greys", origin="upper")

slice_0 = data[91, :, :]
slice_1 = data[:, 109, :]
slice_2 = data[:, :, 91]
show_slices([slice_0, slice_1, slice_2])
plt.suptitle("Center slices for image")
```
